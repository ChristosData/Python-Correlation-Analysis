```python
#Import Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Now we need to read in the data
df = pd.read_csv(r'/Users/home/Downloads/movies.csv')
```

```python
# Looking at the data
df.head()
```

```python
# Checking for incomplete data using a for loop to scan all the data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#drop null values
df = df.dropna()
```

```python
# Data Cleaning - Data Types
df.dtypes
```

```python
# Convert data type for specific column(s)
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
#budget, gross and votes have .0 value removed now as that wasn't needed
df
```

```python
# Creating correct year
# Converting released column into a string to pull data from it, we want the first 4 values starting from the comma delimiter
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
df
```

```python
#Sorting by highest gross values (descending)
# if we want to view all rows we can use pd.set_option('display.max_rows', None)
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Checking for duplicates

df['company'].drop_duplicates().sort_values(ascending=False)

# Drop Duplicates
#df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings (000M)')

plt.xlabel('Budget')

plt.ylabel('Gross Earnings')
plt.show()
```

```python
# plot a regression model on budget vs gross using seaborn

reg = sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"green"}, line_kws={"color":"blue"})

reg.set(xlabel='Budget',
       ylabel='Gross Sales',
       title='Linear Regression: Budget v Gross (000M)')
```

```python
# Correlation analysis
df.corr()
```

```python
#high correlation between budget and gross (0.74)
# Visualise the correlation matrix

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Any cells that have an object type, number them so that we can compare them in a correlation matrix
df_numbered = df

for col_name in df_numbered.columns:
    if(df_numbered[col_name].dtype == 'object'):
        df_numbered[col_name] = df_numbered[col_name].astype('category')
        df_numbered[col_name] = df_numbered[col_name].cat.codes
            
df_numbered
```

```python
#compared to original dataframe
df
```

```python


correlation_matrix = df_numbered.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Pivot a level of the hierarchical index labels

corrmatrix = df_numbered.corr()

corr_pairs = corrmatrix.unstack() 

sortedpairs = corr_pairs.sort_values()

highcorr = sortedpairs[(sortedpairs)>0.5]
highcorr
```
